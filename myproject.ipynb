```python
from pandas_datareader import data, wb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import yfinance as yf
yf.pdr_override()
%matplotlib inline

import plotly
import cufflinks as cf
cf.go_offline()
```

```python
start_date = datetime.datetime(int(2006), int(1), 1)
end_date = datetime.datetime(int(2016), int(1), 1)
```

```python
# Bank Of America
bac = data.get_data_yahoo('BAC', start_date.date(), end_date.date())
bac.head()
```

```python
# CitiGroup
cg = data.get_data_yahoo("C",start_date.date(), end_date.date())
cg.head()
```

```python
# Goldman Sachs
gs = data.get_data_yahoo("GS",start_date.date(), end_date.date())
gs.head()
```

```python
# JP Morgan Chase
jpm = data.get_data_yahoo("JPM",start_date.date(), end_date.date())
jpm.head()
```

```python
# Morgan Stanley
ms = data.get_data_yahoo("MS",start_date.date(), end_date.date())
ms.head()
```

```python
# Wells Fargo
wfc = data.get_data_yahoo("WFC",start_date.date(), end_date.date())
wfc.head()
```

```python
tickers = ["bac", "cg", "gs", "jpm", "ms", "wfc"]
```

```python
bank_stocks = pd.concat([bac, cg, gs, jpm, ms, wfc], axis = 1, keys = tickers)
bank_stocks.head()
```

```python
bank_stocks.columns.names = ["Bank Ticker", "Stock Info"]
bank_stocks.head()
```

```python
for tick in tickers:
    print(tick," ", bank_stocks[tick]["Close"].max())
```

```python
bank_stocks.xs(key = "Close", axis = 1, level = "Stock Info").max()
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick + " Return"] = bank_stocks[tick]["Close"].pct_change()
```
